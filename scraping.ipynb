```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import numpy as np
```

```python
def update(e, w, oe):
    if e > oe:
        if w==1:
            return e + 32/(1+10**((e-oe)/400))
        else:
            return e - 32/(1+10**((oe-e)/400))
    else:
        if w==1:
            return e + 32/(1+10**((e-oe)/400))
        else:
            return e - 32/(1+10**((oe-e)/400))
```

# Getting DataFrame

```python
dfs = []
for school in schools:
    with open("schools_stats/{}.html".format(school)) as f:
        page = f.read()
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_= 'over_header').decompose()
    table = soup.find(id ='all_sgl-basic_NCAAM')
    game_log = pd.read_html(str(table))[0].drop('Unnamed: 2',axis=1).drop(20, axis=0).drop(21, axis=0)
    game_log['School'] = school
    
    dfs.append(game_log)
```

```python
game_logs = pd.concat(dfs)
```

```python
full['eFG%'] = (full['FG'] + 0.5 * full['3P']) / full['FGA']
full['TOV%'] = full['TOV'] / (full['FGA'] + 0.44 * full['FTA'] + full['TOV'])
full['TRB%'] = full['TRB'] / (full['TRB'] + full['TRB.1'])
full['o_eFG%'] = (full['FG.1'] + 0.5 * full['3P.1']) / full['FGA.1']
full['o_TOV%'] = full['TOV.1'] / (full['FGA.1'] + 0.44 * full['FTA.1'] + full['TOV.1'])
full['o_TRB%'] = full['TRB.1'] / (full['TRB.1'] + full['TRB'])
```
